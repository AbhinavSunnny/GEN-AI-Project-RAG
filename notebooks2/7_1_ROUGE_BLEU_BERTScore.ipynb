```python
# ============================================================
# IMPORT LIBRARIES
# ============================================================

import pandas as pd

from rouge_score import rouge_scorer

from nltk.translate.bleu_score import sentence_bleu

from bert_score import score

print(
    "Libraries Loaded Successfully"
)
```

```python
# ============================================================
# LOAD EVALUATION FILES
# ============================================================

dense_gpt4o = pd.read_csv(
    "dense_gpt4o_eval.csv"
)

hybrid_gpt4o = pd.read_csv(
    "hybrid_gpt4o_eval.csv"
)

fusion_gpt4o = pd.read_csv(
    "fusion_gpt4o_eval.csv"
)

dense_llama3 = pd.read_csv(
    "dense_llama3_eval.csv"
)

hybrid_llama3 = pd.read_csv(
    "hybrid_llama3_eval.csv"
)

fusion_llama3 = pd.read_csv(
    "fusion_llama3_eval.csv"
)

dense_mistral = pd.read_csv(
    "dense_mistral_eval.csv"
)

hybrid_mistral = pd.read_csv(
    "hybrid_mistral_eval.csv"
)

fusion_mistral = pd.read_csv(
    "fusion_mistral_eval.csv"
)

print(
    "All Files Loaded Successfully"
)
```

```python
dense_gpt4o = dense_gpt4o.head(25)

hybrid_gpt4o = hybrid_gpt4o.head(25)

fusion_gpt4o = fusion_gpt4o.head(25)

dense_llama3 = dense_llama3.head(25)

hybrid_llama3 = hybrid_llama3.head(25)

fusion_llama3 = fusion_llama3.head(25)

dense_mistral = dense_mistral.head(25)

hybrid_mistral = hybrid_mistral.head(25)

fusion_mistral = fusion_mistral.head(25)
```

```python
# ============================================================
# VERIFY DATA
# ============================================================

print(
    dense_gpt4o.shape
)

print(
    hybrid_gpt4o.shape
)

print(
    fusion_gpt4o.shape
)

print(
    dense_llama3.shape
)

print(
    hybrid_llama3.shape
)

print(
    fusion_llama3.shape
)

print(
    dense_mistral.shape
)

print(
    hybrid_mistral.shape
)

print(
    fusion_mistral.shape
)
```

```python
# ============================================================
# STORE ALL DATAFRAMES
# ============================================================

all_results = {

    "Dense GPT4o":
    dense_gpt4o,

    "Hybrid GPT4o":
    hybrid_gpt4o,

    "Fusion GPT4o":
    fusion_gpt4o,

    "Dense Llama3":
    dense_llama3,

    "Hybrid Llama3":
    hybrid_llama3,

    "Fusion Llama3":
    fusion_llama3,

    "Dense Mistral":
    dense_mistral,

    "Hybrid Mistral":
    hybrid_mistral,

    "Fusion Mistral":
    fusion_mistral

}

print(
    "Dictionary Created"
)
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(
    ["rougeL"],
    use_stemmer=True
)

print(
    "ROUGE Initialized Successfully"
)
```

```python
# ============================================================
# CALCULATE METRICS
# ============================================================

metrics = []

for pipeline, df in all_results.items():

    rouge_scores = []

    bleu_scores = []

    references = []

    predictions = []

    for _, row in df.iterrows():

        reference = str(
            row["reference_summary"]
        )

        prediction = str(
            row["generated_answer"]
        )

        # ------------------------------------
        # ROUGE-L
        # ------------------------------------

        rouge = scorer.score(

            reference,

            prediction

        )["rougeL"].fmeasure

        rouge_scores.append(
            rouge
        )

        # ------------------------------------
        # BLEU
        # ------------------------------------

        bleu = sentence_bleu(

            [reference.split()],

            prediction.split()

        )

        bleu_scores.append(
            bleu
        )

        references.append(
            reference
        )

        predictions.append(
            prediction
        )

    # ----------------------------------------
    # BERTScore
    # ----------------------------------------

    P, R, F1 = score(

        predictions,

        references,

        lang="en",

        verbose=False

    )

    metrics.append({

        "pipeline":
        pipeline,

        "ROUGE_L":
        sum(rouge_scores)

        /

        len(rouge_scores),

        "BLEU":
        sum(bleu_scores)

        /

        len(bleu_scores),

        "BERTScore":
        F1.mean().item()

    })

print(
    "Metrics Calculated"
)
```

```python
# ============================================================
# CREATE METRICS DATAFRAME
# ============================================================

metrics_df = pd.DataFrame(
    metrics
)

metrics_df
```

```python
# ============================================================
# SORT RESULTS
# ============================================================

metrics_df = metrics_df.sort_values(

    by="BERTScore",

    ascending=False

)

metrics_df
```

```python
# ============================================================
# SAVE RESULTS
# ============================================================

metrics_df.to_csv(

    "generation_metrics.csv",

    index=False

)

print(
    "Generation Metrics Saved Successfully"
)
```

The experimental results demonstrate that retrieval strategy significantly impacts the performance of Retrieval-Augmented Generation systems. Among all evaluated pipelines, Fusion GPT4o achieved the highest ROUGE-L, BLEU, and BERTScore values, indicating superior lexical overlap and semantic similarity with reference summaries.

The results further show that Fusion Retrieval consistently outperformed Dense Retrieval across multiple language models, validating the effectiveness of combining dense retrieval, sparse retrieval, and web-based retrieval sources. Mistral-based pipelines achieved competitive performance and, in several cases, outperformed GPT4o configurations using less sophisticated retrieval methods.

Llama3 demonstrated stable performance but generally produced lower evaluation scores compared to GPT4o and Mistral. The findings suggest that retrieval quality contributes substantially to overall answer generation quality and that Fusion Retrieval offers the most robust approach for enhancing RAG systems.
