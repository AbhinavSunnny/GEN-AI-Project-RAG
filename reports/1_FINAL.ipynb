```python
import pandas as pd
```

```python
generation_df = pd.read_csv(

    "../notebooks2/generation_metrics.csv"

)

generation_df
```

```python
retrieval_df = pd.DataFrame({

    "pipeline": [

        "Dense GPT4o",
        "Dense Llama3",
        "Dense Mistral",

        "Fusion GPT4o",
        "Fusion Llama3",
        "Fusion Mistral",

        "Hybrid GPT4o",
        "Hybrid Llama3",
        "Hybrid Mistral"

    ],

    "latency_seconds": [

        2.37,
        27.12,
        34.92,

        3.09,
        43.53,
        55.97,

        2.08,
        26.12,
        37.48

    ],

    "memory_mb": [

        0.03,
        46.06,
        6.81,

        75.11,
        -8.75,
        -5.07,

        8.35,
        24.37,
        -1.61

    ],

    "context_length": [

        2143.35,
        2143.35,
        2143.20,

        4541.65,
        4615.55,
        4714.35,

        2226.05,
        2226.05,
        2226.20

    ],

    "response_length": [

        673.95,
        701.80,
        668.15,

        734.80,
        947.45,
        791.05,

        550.85,
        823.65,
        679.65

    ],

    "context_relevance": [

        0.5622,
        0.5622,
        0.5615,

        0.5417,
        0.5390,
        0.5464,

        0.5587,
        0.5587,
        0.5589

    ]

})

retrieval_df
```

```python
final_df = generation_df.merge(

    retrieval_df,

    on="pipeline",

    how="inner"

)

final_df
```

```python
final_df["Overall_Rank"] = (

    final_df["BERTScore"]

    .rank(

        ascending=False,

        method="dense"

    )

)

final_df = final_df.sort_values(

    by="Overall_Rank"

)

final_df
```

```python
final_df.to_csv(

    "Final_Project_Results.csv",

    index=False

)

print(
    "CSV Saved Successfully"
)
```
