```python
# ============================================================
# IMPORT LIBRARIES
# ============================================================

import pandas as pd

print(
    "Libraries Loaded Successfully"
)
```

```python
# ============================================================
# LOAD CNN DATASET
# ============================================================

df = pd.read_csv(
    "cnn_processed20k.csv"
)

print(
    "Total Rows:",
    len(df)
)

df.head()
```

```python
# ============================================================
# CREATE 1000 ROW EVALUATION SAMPLE
# ============================================================

evaluation_df = df.sample(

    n=1000,

    random_state=42

)

evaluation_df = evaluation_df.reset_index(
    drop=True
)

print(
    "Evaluation Rows:",
    len(evaluation_df)
)
```

```python
# ============================================================
# CREATE QUERY COLUMN
# ============================================================

def create_query(summary):

    words = str(summary).split()

    return " ".join(
        words[:8]
    )

evaluation_df[
    "query"
] = evaluation_df[
    "clean_summary"
].apply(
    create_query
)

print(
    "Queries Created Successfully"
)
```

```python
# ============================================================
# CREATE FINAL EVALUATION DATASET
# ============================================================

final_eval_df = evaluation_df[

    [

        "query",

        "clean_article",

        "clean_summary"

    ]

].copy()

final_eval_df.columns = [

    "query",

    "article",

    "reference_summary"

]

print(
    final_eval_df.shape
)

final_eval_df.head()
```

```python
# ============================================================
# CHECK FOR NULL VALUES
# ============================================================

print(
    final_eval_df.isnull().sum()
)
```

```python
# ============================================================
# SAVE EVALUATION DATASET
# ============================================================

final_eval_df.to_csv(

    "evaluation_dataset.csv",

    index=False

)

print(
    "Evaluation Dataset Saved Successfully"
)
```
