```python
# ============================================================
# IMPORT LIBRARIES
# ============================================================

import pandas as pd

from langchain_text_splitters import (
    RecursiveCharacterTextSplitter
)
```

```python
# ============================================================
# LOAD DATASET
# ============================================================

cnn_df = pd.read_csv(
    "../notebooks/cnn_processed20k.csv"
)

print(
    "Dataset Shape:",
    cnn_df.shape
)
```

```python
# ============================================================
# CHUNK CONFIGURATIONS
# ============================================================

chunk_configs = [

    (250, 25),

    (500, 50),

    (750, 75),

    (1000, 100)
]
```

```python
# ============================================================
# CHUNKING EXPERIMENT
# ============================================================

results = []

for chunk_size, overlap in chunk_configs:

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=overlap
    )

    all_chunks = []

    for article in cnn_df[
        "clean_article"
    ].head(1000):

        chunks = splitter.split_text(
            str(article)
        )

        all_chunks.extend(
            chunks
        )

    results.append(
        {
            "Chunk Size": chunk_size,
            "Overlap": overlap,
            "Total Chunks": len(all_chunks),
            "Chunks Per Article":
            round(
                len(all_chunks) / 1000,
                2
            )
        }
    )
```

```python
# ============================================================
# VIEW RESULTS
# ============================================================

pd.DataFrame(
    results
)
```
